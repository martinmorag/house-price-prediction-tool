# House Price Prediction - Data Cleaning & Preprocessing

## Objective

The objective of this notebook is to prepare the dataset for machine learning by addressing data quality issues identified during the exploratory data analysis (EDA).

The preprocessing steps performed in this notebook include:

- Handling missing values
- Verifying duplicate records
- Detecting potential outliers
- Preparing features for machine learning
- Saving a clean dataset for the modeling stage

By the end of this notebook, the dataset will be ready for feature engineering and model training.

```python
import pandas as pd
import numpy as np

from sklearn.impute import SimpleImputer
```

## 1. Load the Dataset

The cleaned dataset will be created from the original raw data. Before applying any preprocessing techniques, we load the dataset and inspect its basic structure.

```python
house_data = pd.read_csv("../data/raw/dataset_2.csv")

house_data.head()
```

## 2. Dataset Overview

Before modifying the data, it is important to verify the size of the dataset, the data types of each feature, and the presence of missing values.

```python
print(f"Dataset Shape: {house_data.shape}")

house_data.info()
```

## 3. Missing Values Analysis

During the exploratory data analysis, we identified missing values in several columns.

Before training a machine learning model, these missing values must be handled appropriately to avoid errors and improve model performance.

```python
house_data.isnull().sum()
```

### Numerical Missing Values

The numerical features **Area_SqFt** and **Rooms** contain missing values.

Instead of removing these observations, the missing values will be replaced using the median.

The median is preferred because it is less affected by extreme values than the mean, making it a more robust measure for real estate data.

```python
numeric_columns = ["Area_SqFt", "Rooms"]

median_imputer = SimpleImputer(strategy="median")

house_data[numeric_columns] = median_imputer.fit_transform(
    house_data[numeric_columns]
)
```

### Categorical Missing Values

The **Furnishing** feature also contains missing values.

Since this is a categorical variable, the missing entries will be replaced with the most frequent category (mode). This approach preserves the distribution of the feature while avoiding the loss of observations.

```python
categorical_columns = ["Furnishing"]

mode_imputer = SimpleImputer(strategy="most_frequent")

house_data[categorical_columns] = mode_imputer.fit_transform(
    house_data[categorical_columns]
)
```

## 4. Verify Missing Values

After applying the imputation techniques, we verify that all missing values have been successfully handled.

```python
house_data.isnull().sum()
```

## 5. Duplicate Records

Duplicate records may introduce bias into a machine learning model by giving excessive importance to repeated observations.

We check whether duplicate rows exist before continuing with the preprocessing pipeline.

```python
duplicates = house_data.duplicated().sum()

print(f"Duplicate rows: {duplicates}")
```

## 6. Summary

The following preprocessing steps have been completed:

- Missing numerical values were replaced using the median.
- Missing categorical values were replaced using the most frequent category.
- The dataset was verified to contain no remaining missing values.
- Duplicate records were checked.

The dataset is now ready for outlier analysis and feature preprocessing.
