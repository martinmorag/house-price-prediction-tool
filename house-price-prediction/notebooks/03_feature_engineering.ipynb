# House Price Prediction - Feature Engineering

## Objective

The purpose of this notebook is to prepare the dataset for machine learning by transforming the raw features into a format that regression models can understand.

The main tasks performed in this notebook are:

- Separate the features and target variable
- Split the dataset into training and testing sets
- Identify numerical and categorical features
- Build a preprocessing pipeline
- Encode categorical variables
- Scale numerical variables
- Create a reusable machine learning pipeline

Using a preprocessing pipeline ensures that all transformations are applied consistently during both training and prediction while preventing data leakage.

```python
import pandas as pd

from sklearn.model_selection import train_test_split

from sklearn.compose import ColumnTransformer

from sklearn.pipeline import Pipeline

from sklearn.preprocessing import (
    OneHotEncoder,
    StandardScaler
)

from sklearn.impute import SimpleImputer
```

## 1. Load the Clean Dataset

The cleaned dataset from the previous notebook is loaded before applying feature engineering techniques.

```python
house_data = pd.read_csv("../data/raw/dataset_2.csv")

# Numerical imputation
house_data[["Area_SqFt", "Rooms"]] = (
    SimpleImputer(strategy="median")
    .fit_transform(house_data[["Area_SqFt", "Rooms"]])
)

# Categorical imputation
house_data[["Furnishing"]] = (
    SimpleImputer(strategy="most_frequent")
    .fit_transform(house_data[["Furnishing"]])
)

house_data.head()
```

## 2. Separate Features and Target

Machine learning models require separating the independent variables (features) from the dependent variable (target).

The target variable for this project is **Price**, while all remaining columns will be used as predictors.

```python
X = house_data.drop("Price", axis=1)

y = house_data["Price"]

print("Features shape:", X.shape)
print("Target shape:", y.shape)
```

## 3. Split the Dataset

To evaluate the model fairly, the dataset is divided into training and testing sets.

The model will learn patterns using the training data and will be evaluated on unseen testing data.

An 80/20 split is commonly used for supervised learning problems.

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42
)

print("Training samples:", X_train.shape[0])
print("Testing samples:", X_test.shape[0])
```

## 4. Identify Feature Types

Different preprocessing techniques are required for numerical and categorical features.

Numerical variables will be scaled, while categorical variables will be encoded using One-Hot Encoding.

```python
numerical_features = [
    "Area_SqFt",
    "Rooms",
    "Build_Year"
]

categorical_features = [
    "Location",
    "Street_Type",
    "Furnishing",
    "Property_Type",
    "Has_Pool"
]

print("Numerical Features")
print(numerical_features)

print()

print("Categorical Features")
print(categorical_features)
```

## 5. Create Preprocessing Pipelines

A preprocessing pipeline is used to automate all transformations applied to the data.

For numerical variables:

- Missing values are replaced using the median.
- Features are standardized using StandardScaler.

For categorical variables:

- Missing values are replaced with the most frequent value.
- Categories are converted into numerical features using One-Hot Encoding.

```python
numeric_transformer = Pipeline(
    steps=[
        ("imputer", SimpleImputer(strategy="median")),
        ("scaler", StandardScaler())
    ]
)

categorical_transformer = Pipeline(
    steps=[
        ("imputer", SimpleImputer(strategy="most_frequent")),
        (
            "encoder",
            OneHotEncoder(handle_unknown="ignore")
        )
    ]
)
```

## 6. Combine the Pipelines

The numerical and categorical preprocessing pipelines are combined into a single ColumnTransformer.

This ensures that each feature receives the appropriate preprocessing before being passed to the machine learning model.

```python
preprocessor = ColumnTransformer(
    transformers=[
        (
            "num",
            numeric_transformer,
            numerical_features
        ),
        (
            "cat",
            categorical_transformer,
            categorical_features
        )
    ]
)

preprocessor
```

## 7. Summary

The dataset is now fully prepared for machine learning.

The preprocessing pipeline automatically:

- Handles missing values
- Scales numerical features
- Encodes categorical features
- Prevents data leakage
- Ensures consistent transformations during training and prediction

In the next notebook, this preprocessing pipeline will be combined with several regression models to compare their predictive performance.
