# House Price Prediction - Model Interpretation

## Objective

Building an accurate model is only one part of a successful machine learning project. Equally important is understanding how the model makes its predictions.

In this notebook, the trained Random Forest model is analyzed to identify the features that have the greatest influence on house prices.

Understanding feature importance provides valuable business insights and helps explain the model's behavior to both technical and non-technical stakeholders.

```python
import joblib
import pandas as pd
import matplotlib.pyplot as plt
```

## 1. Load the Trained Model

The optimized machine learning pipeline created in the previous notebook is loaded from disk.

Because the preprocessing steps and the regression model were saved together as a single pipeline, no additional preprocessing is required.

```python
model = joblib.load("../models/best_model.pkl")
```

## 2. Load the Dataset

The dataset is loaded so that feature names can be reconstructed after preprocessing.

This allows the feature importance values produced by the Random Forest model to be associated with meaningful feature names.

```python
house_data = pd.read_csv("../data/raw/dataset_2.csv")

house_data.head()
```

## 3. Retrieve the Trained Random Forest Model

The saved pipeline consists of two main components:

- The preprocessing pipeline
- The trained Random Forest model

The regression model is extracted so that its feature importance values can be examined.

```python
random_forest = model.named_steps["model"]

preprocessor = model.named_steps["preprocessor"]
```

## 4. Recover Feature Names

Categorical variables are transformed using One-Hot Encoding.

After encoding, a single categorical feature (such as `Location`) becomes multiple binary columns.

The following code reconstructs the transformed feature names so they can be displayed alongside their importance scores.

```python
feature_names = preprocessor.get_feature_names_out()
```

```python
feature_importance = pd.DataFrame({
    "Feature": feature_names,
    "Importance": random_forest.feature_importances_
})

feature_importance = feature_importance.sort_values(
    by="Importance",
    ascending=False
)

feature_importance.head(10)
```

## 5. Visualize Feature Importance

Feature importance provides an estimate of how much each variable contributes to the model's predictions.

Higher importance values indicate that a feature plays a greater role in determining house prices.

```python
top_features = feature_importance.head(10)

plt.figure(figsize=(10,6))

plt.barh(
    top_features["Feature"],
    top_features["Importance"]
)

plt.title("Top 10 Most Important Features")

plt.xlabel("Importance")

plt.gca().invert_yaxis()

plt.show()
```

## 6. Interpretation of Results

The feature importance analysis indicates which property characteristics have the greatest influence on predicted house prices.

If `Area_SqFt` appears as the most important feature, this suggests that property size is the strongest predictor of price within this dataset.

Categorical variables such as location, property type, or the presence of a swimming pool may also contribute significantly, reflecting the influence of neighborhood and amenities on property values.

Although feature importance indicates the relative contribution of each variable, it does not imply causation. Instead, it reflects how much the model relied on each feature when making predictions.
